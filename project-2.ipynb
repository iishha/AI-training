```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[2, 0], [0, 1]])

grid = np.array([[x, y] for x in range(-5, 6) for y in range(-5, 6)]).T

grid_transformed = A @ grid

plt.scatter(*grid, label='Original Grid')
plt.scatter(*grid_transformed, label='Transformed Grid')

plt.legend()
plt.show()
```
