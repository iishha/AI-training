```python
#Organizational hierarchy

class Node():
    def __init__(self, val): 
        self.val = val
        self.left = None
        self.right = None


root = Node("CEO")
root.left = Node("Manager1")
root.right = Node("Manager2")
root.left.left = Node("Employee1")
root.right.right = Node("Employee2")


def inorder(root):
    if root:
        inorder(root.left)
        print(root.val, end=" ")
        inorder(root.right)

def preorder(root):
    if root:
        print(root.val, end=" ")
        preorder(root.left)
        preorder(root.right)

def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.val, end=" ")

def print_tree(node, level=0):
    if node is not None:
        print_tree(node.right, level + 1)
        print("       " * level + str(node.val))
        print_tree(node.left, level + 1)


print_tree(root)
```

```python
# 2. Tournament Bracket System

class Node():
    def __init__(self, val):   
        self.val = val
        self.left = None
        self.right = None

q1 = Node("Q1")
q2 = Node("Q2")
q3 = Node("Q3")
q4 = Node("Q4")

semi1 = Node("Semi1")
semi1.left = q1
semi1.right = q2

semi2 = Node("Semi2")
semi2.left = q3
semi2.right = q4
final = Node("Final")
final.left = semi1
final.right = semi2
def inorder(root):
    if root:
        inorder(root.left)
        print(root.val, end=" ")
        inorder(root.right)

def preorder(root):
    if root:
        print(root.val, end=" ")
        preorder(root.left)
        preorder(root.right)

def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.val, end=" ")

def print_tree(node, level=0):
    if node:
        print_tree(node.right, level + 1)
        print("    " * level + node.val)
        print_tree(node.left, level + 1)
print_tree(final)
print("\n\nInorder Traversal:")
inorder(final)

print("\n\nPreorder Traversal:")
preorder(final)

print("\n\nPostorder Traversal:")
postorder(final)
```

```python
#Expression Evaluation in compilers
class Node:
    def __init__(self, val): 
        self.val = val
        self.left = None
        self.right = None

root = Node("*")
root.left = Node("+")
root.right = Node("-")

root.left.left = Node("3")
root.left.right = Node("5")
root.right.left = Node("2")
root.right.right = Node("1")

def inorder(root):
    if root:
        if root.left or root.right:
            print("(", end="")
        inorder(root.left)
        print(root.val, end=" ")
        inorder(root.right)
        if root.left or root.right:
            print(")", end="")
def preorder(root):
    if root:
        print(root.val, end=" ")
        preorder(root.left)
        preorder(root.right)

def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.val, end=" ")

def evaluate(root):
    if not root.left and not root.right:
        return int(root.val)
    l = evaluate(root.left)
    r = evaluate(root.right)
    if root.val == '+': return l + r
    if root.val == '-': return l - r
    if root.val == '*': return l * r
    if root.val == '/': return l // r

def print_tree(node, level=0):
    if node:
        print_tree(node.right, level + 1)
        print("    " * level + node.val)
        print_tree(node.left, level + 1)

print("Tree Structure:")
print_tree(root)

print("\nInorder Expression:")
inorder(root)

print("\n\nPreorder Expression:")
preorder(root)

print("\n\nPostorder Expression:")
postorder(root)

print("\n\nEvaluated Result:")
print(evaluate(root))
```
